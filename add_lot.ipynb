```python
import numpy as np
import znh5md
from ase.io import read
```

```python
b3lyp = read("data/buoh/buoh_b3lyp.traj", ":")
xtb = read("data/buoh/buoh_xtb.traj", ":")
```

```python
lot0 = np.array([1,0])
lot1 = np.array([1,1])
```

```python
for a in xtb:
    a.info["fidelity"] = lot0

for a in b3lyp:
    a.info["fidelity"] = lot1
```

```python
all_atoms = xtb + b3lyp
```

```python
znh5md.write("buoh_mf.h5", all_atoms)
```
